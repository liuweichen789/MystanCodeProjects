```python
# prompt: 請用leadingIndicatorTaiwan資料，年月為x軸，數值為y軸，繪製景氣信號燈圖，y軸最高為60，並用實心圓點標記數值，數值低於16以藍色標記，市值介於22-17的數值用黃藍色標記，數值介於31-23的數值用綠色標記，數值介於37-32用橘紅色標記，數值大於38用紅色標記

import pandas as pd
import matplotlib.pyplot as plt

try:
    leadingIndicatorTaiwan = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/TEJ/EA1101 台灣-景氣對策信號綜合分數2025.csv")
    # Assuming the date column is named '年月' and the value column is named '數值'
    # Adjust column names if necessary
    plt.figure(figsize=(12, 6))

    # plot data points with different color based on value
    for index, row in leadingIndicatorTaiwan.iterrows():
      value = row['數值']
      color = 'blue'
      if value < 16:
        color = 'blue'
      elif 17 <= value <= 22:
          color = 'yellow' # yellow
      elif 23 <= value <= 31:
          color = 'green'
      elif 32 <= value <= 37:
          color = 'orange'
      elif value >=38:
          color = 'red'
      plt.plot(row['年月'], value, marker='o', markersize=8, color=color)

    plt.xlabel('年月')
    plt.ylabel('數值')
    plt.title('景氣信號燈圖')
    plt.ylim(0, 60) # set y axis limit
    plt.xticks(rotation=45, ha='right')
    plt.grid(True)
    plt.tight_layout()
    plt.show()

except FileNotFoundError:
    print("錯誤：找不到檔案。請確認檔案路徑是否正確。")
except KeyError as e:
  print(f"錯誤: 資料欄位不存在。請確認 '年月' 和 '數值' 欄位是否存在於 CSV 檔案中。錯誤訊息: {e}")
except Exception as e:
    print(f"發生未知錯誤：{e}")
```
