```python
# prompt: 請依照df_name資料，依照股票名稱個別計算K值與D值，另存成df_KD

import pandas as pd
# Calculate K and D values for each stock
df_KD = pd.DataFrame()
for stock_name in df_name['股票名稱'].unique():
    df_temp = df_name[df_name['股票名稱'] == stock_name].copy()
    # Calculate K and D using talib
    df_temp['K'], df_temp['D'] = talib.STOCH(df_temp['High'], df_temp['Low'], df_temp['Close'], fastk_period=9, slowk_period=3, slowk_matype=0, slowd_period=3, slowd_matype=0)
    df_KD = pd.concat([df_KD, df_temp])

# Display the DataFrame with K and D values
df_KD
```

```python
# prompt: 請依照df_cal資料，依照股票名稱個別計算RSI, MACD, MACD_signal, MACD_hist, MA20乖離率, MA20, upperband, middleban,d, lowerband，另存成df_cal2

import pandas as pd
# Assuming df_cal is already defined as in the provided code.

df_cal2 = pd.DataFrame()
for stock_name in df_cal['股票名稱'].unique():
    df_temp = df_cal[df_cal['股票名稱'] == stock_name].copy()
    # Calculate RSI
    df_temp['RSI'] = talib.RSI(df_temp['Adj Close'], timeperiod=14)

    # Calculate MACD
    df_temp['MACD'], df_temp['MACD_signal'], df_temp['MACD_hist'] = talib.MACD(df_temp['Adj Close'], fastperiod=12, slowperiod=26, signalperiod=9)

    # Calculate MA20 and MA20乖離率
    df_temp['MA20'] = talib.SMA(df_temp['Adj Close'], timeperiod=20)
    df_temp['MA20乖離率'] = (df_temp['Adj Close'] - df_temp['MA20']) / df_temp['MA20']

    # Calculate Bollinger Bands
    df_temp['upperband'], df_temp['middleband'], df_temp['lowerband'] = talib.BBANDS(df_temp['Adj Close'], timeperiod=20, nbdevup=2, nbdevdn=2, matype=0)

    df_cal2 = pd.concat([df_cal2, df_temp])

# Display the DataFrame with calculated indicators
df_cal2
```

```python
# prompt: 請依照df_cal4MA20_macd2資料，依照股票名稱分別計算，擷取第n天buy為1，第n-1天buy為0的資料，擷取第n天sell為1，第n-1天sell為0的資料

import pandas as pd
df_buy = pd.DataFrame()
df_sell = pd.DataFrame()

for stock_name in df_cal4MA20_macd2['股票名稱'].unique():
    df_temp = df_cal4MA20_macd2[df_cal4MA20_macd2['股票名稱'] == stock_name].copy()
    # 找出buy訊號(當天buy為1，前一天buy為0)
    df_temp['buy_signal'] = ((df_temp['buy'] == 1) & (df_temp['buy'].shift(1) == 0))
    df_buy = pd.concat([df_buy, df_temp[df_temp['buy_signal']]])
    # 找出sell訊號(當天sell為1，前一天sell為0)
    df_temp['sell_signal'] = ((df_temp['sell'] == 1) & (df_temp['sell'].shift(1) == 0))
    df_sell = pd.concat([df_sell, df_temp[df_temp['sell_signal']]])

# 顯示buy訊號資料
print("Buy Signals:")
print(df_buy)

# 顯示sell訊號資料
print("\nSell Signals:")
df_sell


# Save df_buy and df_sell to Excel files if needed
# df_buy.to_excel('/content/drive/MyDrive/Colab Notebooks/data/df_buy.xlsx', index=True)
# df_sell.to_excel('/content/drive/MyDrive/Colab Notebooks/data/df_sell.xlsx', index=True)
```

```python
# prompt: 請結合Buy Signals與Sell Signals

import pandas as pd
df_signals = pd.DataFrame()

for stock_name in df_cal4MA20_macd2['股票名稱'].unique():
    df_temp = df_cal4MA20_macd2[df_cal4MA20_macd2['股票名稱'] == stock_name].copy()

    # 找出買進訊號 (當天買進為1，前一天買進為0，且 MACD 買進為1)
    df_temp['buy_signal'] = ((df_temp['buy'] == 1) & (df_temp['buy'].shift(1) == 0) & (df_temp['macd_buy'] == 1))
    # 找出賣出訊號 (當天賣出為1，前一天賣出為0，且 MACD 賣出為1)
    df_temp['sell_signal'] = ((df_temp['sell'] == 1) & (df_temp['sell'].shift(1) == 0) & (df_temp['macd_sell'] == 1))

    df_signals = pd.concat([df_signals, df_temp])

# 顯示包含買賣訊號的資料
print("Combined Buy and Sell Signals:")
df_signals

# Save df_signals to an Excel file if needed
df_signals.to_excel('/content/drive/MyDrive/Colab Notebooks/data/df_signals.xlsx', index=True)
```

```python
# prompt: 請依照df_signals資料，依照股票名稱個別計算，當buy_signal為True，以Adj Close值買進，增加1個欄位為買進價格，當sell_signal為True，以Adj Close值賣出，增加1個欄位為賣出價格，另存為df_signals2

import pandas as pd
import numpy as np
df_signals2 = df_signals.copy()

for stock_name in df_signals2['股票名稱'].unique():
  df_temp = df_signals2[df_signals2['股票名稱'] == stock_name].copy()

  # 買進價格
  df_temp['買進價格'] = np.where(df_temp['buy_signal'], df_temp['Adj Close'], np.nan)

  # 賣出價格
  df_temp['賣出價格'] = np.where(df_temp['sell_signal'], df_temp['Adj Close'], np.nan)

  df_signals2 = pd.concat([df_signals2[df_signals2['股票名稱'] != stock_name], df_temp])

# 顯示包含買賣價格的資料
print("Combined Buy and Sell Prices:")
df_signals2
```

```python
# prompt: 請依照df_ma60_macd5資料，依照股票名稱個別計算，分別加總買進價格，加總賣出價格，另存成df_ma60_macd6，並以表格顯示

# 根據 df_ma60_macd5 資料，依照股票名稱個別計算，分別加總買進價格，加總賣出價格，另存成 df_ma60_macd6，並以表格顯示
df_ma60_macd6 = df_ma60_macd5.groupby('股票名稱').agg({'買進價格': 'sum', '賣出價格': 'sum'})

# 顯示 df_ma60_macd6 表格
print("df_ma60_macd6 表格：")
df_ma60_macd6
```

```python
# prompt: 請依照df_ma60_macd6資料，新增一個欄位為損益，依照股票名稱個別計算，將買進價格-賣出價格的值填入

# 複製一份 df_ma60_macd6 資料
df_ma60_macd6_profit = df_ma60_macd6.copy()

# 新增 '損益' 欄位，並計算買進價格減賣出價格
df_ma60_macd6_profit['損益'] = df_ma60_macd6_profit['買進價格'] - df_ma60_macd6_profit['賣出價格']

# 顯示包含損益的 df_ma60_macd6_profit 表格
print("包含損益的 df_ma60_macd6 表格：")
df_ma60_macd6_profit
```

```python
# prompt: 請依照df_ma60_macd6資料，新增一個欄位為百分比，依照股票名稱個別計算，將損益/買進價格*100，取小數點後1位，其餘四捨五入計算後的值填入

import numpy as np
# 複製一份 df_ma60_macd6 資料
df_ma60_macd6_percentage = df_ma60_macd6.copy()

# 新增 '百分比' 欄位，並初始化為 NaN
df_ma60_macd6_percentage['百分比'] = np.nan

# 遍歷 df_ma60_macd6_profit 中的股票名稱和損益
for stock_name, row in df_ma60_macd6_profit.iterrows():
  # 計算百分比，並四捨五入到小數點後一位
  if row['買進價格'] != 0:
    percentage = (row['損益'] / row['買進價格'] * 100).round(1)
    # 將百分比填入 '百分比' 欄位
    df_ma60_macd6_percentage.loc[stock_name, '百分比'] = percentage

# 顯示包含百分比的 df_ma60_macd6_percentage 表格
print("包含百分比的 df_ma60_macd6 表格：")
df_ma60_macd6_percentage
```
