# 1.下載 ETF0050 五十檔 2010-01-01 到今天的歷史交易資料

(含tickers, Date, Open, High, Low, Close, AdjClose, Volume)

```python
# prompt: pd.read('/content/drive/MyDrive/Colab Notebooks/data/股票/Taiwan50ETF.xlsx')

dfName = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/data/股票/Taiwan50ETF.xlsx')
dfName
```

```python
dfName["股票代碼"]
```

```python
taiwan_50_tickers = dfName["股票代碼"].to_list()
taiwan_50_tickers
```

```python
# prompt: taiwan_50_tickers 轉成大寫

taiwan_50_tickers_upper = [ticker.upper() for ticker in taiwan_50_tickers]
taiwan_50_tickers_upper
```

```python
# prompt: 從" yfinance " 中 台灣50ETF 成分股 2010-01-01 到2025-03-30 資料 欄位為資料欄位為順序為"tickers",'Date', 'Open', 'High', 'Low', 'Close', 'Volume'

import pandas as pd
import yfinance as yf
'''
# 台灣50 ETF (0050.TW) 成分股代碼（台股需加 .TW）
taiwan_50_tickers = [
    "2330.TW", "2454.TW", "2317.TW", "6505.TW", "2412.TW", "1303.TW", "2882.TW", "1301.TW", "2881.TW", "2002.TW",
    "5871.TW", "2308.TW", "3008.TW", "2886.TW", "2891.TW", "2884.TW", "2885.TW", "1101.TW", "1326.TW", "2892.TW",
    "9910.TW", "2880.TW", "3711.TW", "2207.TW", "2603.TW", "5876.TW", "1216.TW", "2609.TW", "3034.TW", "2327.TW",
    "1590.TW", "2382.TW", "4904.TW", "2395.TW", "6531.TW", "4938.TW", "1402.TW", "2606.TW", "2912.TW", "6415.TW",
    "3045.TW", "2890.TW", "2379.TW", "2883.TW", "2408.TW", "2303.TW", "6504.TW", "3702.TW", "2357.TW", "1210.TW"
]
'''
# 設定歷史資料範圍
start_date = "2010-01-01"
end_date = datetime.datetime.now()

# 抓取歷史股價資料
data = yf.download(taiwan_50_tickers_upper, start=start_date, end=end_date, group_by='ticker',auto_adjust= False)

# 處理資料，將資料重新組織成需要的格式
result_data = []
for ticker in taiwan_50_tickers_upper:
    if ticker in data:
        df = data[ticker]
        df['tickers'] = ticker  # 新增 'tickers' 欄位
        df = df.reset_index()  # 將日期索引轉為欄位
        result_data.append(df)

# 合併所有股票的資料
result_df = pd.concat(result_data)

# 指定欄位順序
result_df = result_df[['tickers', 'Date', 'Open', 'High', 'Low', 'Close','Adj Close', 'Volume']]

# 顯示結果
print(result_df)

#儲存檔案
result_df.to_csv('/content/drive/MyDrive/Colab Notebooks/data/股票/taiwan50_2010_2025Adj.csv', index=False)
# 9秒
```
